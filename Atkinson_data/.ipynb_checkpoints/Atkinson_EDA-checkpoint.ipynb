```python
import pandas as pd
import os
import re
import shutil
from multiprocessing.pool import ThreadPool
```

```python
def copy_file(src_path, dst_path):
    shutil.copy(src_path, dst_path)

def copy_files(files_list, src_dir, dst_dir, num_threads=4):
    src_paths = [os.path.join(src_dir, f) for f in files_list]
    dst_paths = [os.path.join(dst_dir, f) for f in files_list]
    pool = ThreadPool(num_threads)
    pool.starmap(copy_file, zip(src_paths, dst_paths))
    pool.close()
    pool.join()
```

```python
# LAB data import 
lab_df = pd.read_csv("F:\Beetle_classification_deep_data_segmented_clean\segmented_images_metadata.csv", index_col="Unnamed: 0")
lab_df
```

```python
# Atkinson data
atkinson_df = pd.read_csv("F:\Atkinson_data\images_metadata.csv")
atkinson_df
```

```python
# get all files aned paths 
# path = r'F:\Atkinson_data\hand_selected_images'
path = r'F:\Atkinson_data\script_selected_images'
all_paths = []
all_files = []
for dirpath, dirnames, filenames in os.walk(path):
    for filename in filenames:
        if not filename.startswith("._"):
            if filename.endswith(".jpg"):
                all_paths.append(os.path.join(dirpath, filename))
                all_files.append(filename[:-4])
```

```python
# filter CSV by files found
atkinson_df_selected = atkinson_df[atkinson_df['file_name'].isin(all_files)]
print("Number of images not in CSV: ", len(all_files) - len(atkinson_df_selected))
```

```python
atkinson_df_selected["genus_species"] = atkinson_df_selected['_1 Valid Names::genus']+'_'+atkinson_df_selected['_1 Valid Names::species']
spcs_lst = atkinson_df_selected["genus_species"].unique().tolist()
atkinson_df_selected["full_file_name"] = atkinson_df_selected["file_name"]+".jpg"
all_images_lst = atkinson_df_selected["full_file_name"].unique().tolist()
atkinson_df_selected
```

```python
# save data inot new folder
target_dir = r'F:\Atkinson_data\images_copy'
source_dir = r'F:\Atkinson_data\script_selected_images'
atkinson_df_selected.to_csv(target_dir+"\metadata.csv")
copy_files(files_list=all_images_lst, src_dir= source_dir, dst_dir=target_dir, num_threads=4)
```
